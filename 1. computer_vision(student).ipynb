# How to measure how bad our model is??

```python
def mse(y_true, y_pred):
    error = y_true - y_pred
    sq_err = np.power(error,2)
    return np.mean(sq_err)

mse(y,y_hat)
```

# Linear regression with pytorch

```python
import torch
from torch import tensor,nn
```

```python
n=100
```

```python
x = torch.ones(n,2) 
x[:,0].uniform_(-1.,1)
x[:5]
```

```python
a = tensor([3.,2]); a
```

```python
y = x@a + torch.rand(n)
```

```python
plt.scatter(x[:,0], y);
```

```python
def mse(y_hat, y): return ((y_hat-y)**2).mean()
```

```python
wb = tensor([-1.,1])
```

```python
y_hat = x@wb
mse(y_hat, y)
```

```python
plt.scatter(x[:,0],y)
plt.scatter(x[:,0],y_hat);
```

```python
wb = nn.Parameter(wb); wb
```

```python
def update():
    y_hat = x@wb
    loss = mse(y, y_hat)
    if t % 10 == 0: print(loss)
    loss.backward()
    with torch.no_grad():
        wb.sub_(lr * wb.grad)
        wb.grad.zero_()
```

```python
lr = 1e-1
for t in range(100): update()
```

```python
plt.scatter(x[:,0],y)
y_hat = x@wb
plt.scatter(x[:,0],y_hat.detach().numpy());
```

```python
from matplotlib import animation, rc
rc('animation', html='jshtml')
```

```python
wb = nn.Parameter(tensor([-1.,1]))

fig = plt.figure()
plt.scatter(x[:,0], y, c='orange')
y_hat = x@wb

line, = plt.plot(x[:,0].detach().numpy(),y_hat.detach().numpy())
plt.close()

def animate(i):
    update()
    y_hat = x@wb
    line.set_ydata(y_hat.detach().numpy())
    return line,

animation.FuncAnimation(fig, animate, np.arange(0, 100), interval=20)
```
